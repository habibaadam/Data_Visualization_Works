```
#PieChart representing the best selling phone manufacturers according to phones they produced and ranks from 2000-2022

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#Importing Best Selling Phones DataSet
BestSellingP =pd.read_csv('bestsellingphones.csv')

#To Determine the column names.
BestSellingP.columns
```

```
#Counts of times specific manufacturers produced phones and ranks from 2000-2022

Apple_Phones= BestSellingP.loc[BestSellingP['Manufacturer']=='Apple'].count()[0]
Samsung_Phones= BestSellingP.loc[BestSellingP['Manufacturer']=='Samsung'].count()[0]
Nokia_Phones= BestSellingP.loc[BestSellingP['Manufacturer']=='Nokia'].count()[0]
Huawei_Phones= BestSellingP.loc[BestSellingP['Manufacturer']=='Huawei'].count()[0]
LG_Phones= BestSellingP.loc[BestSellingP['Manufacturer']=='LG'].count()[0]
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(7,7))
Apple_Phones= BestSellingP.loc[BestSellingP['Manufacturer']=='Apple'].count()[0]
Nokia_Phones= BestSellingP.loc[BestSellingP['Manufacturer']=='Nokia'].count()[0]
Samsung_Phones= BestSellingP.loc[BestSellingP['Manufacturer']=='Samsung'].count()[0]
Huawei_Phones= BestSellingP.loc[BestSellingP['Manufacturer']=='Huawei'].count()[0]
LG_Phones= BestSellingP.loc[BestSellingP['Manufacturer']=='LG'].count()[0]
labels=["Apple Phones","Nokia Phones","Samsung Phones","Huawei Phones","LG Phones"]
colors=['m','y','g','r','b']
plt.title('Manufacturer Productions Throughout The Years',rotation = 360)
plt.pie([Apple_Phones,Nokia_Phones,Samsung_Phones,Huawei_Phones,LG_Phones],labels=labels,colors=colors,autopct='%.1f%%')
plt.legend()
plt.show()

#From the pie chart plotted below,the samsung manufacturer appeared most frequent in the ranks from 200-2022 the most.
#Huawei manufacturer produced the least phones
```

```
#Pie Chart Representing Percentages Of The Form Factors(Touchscreen/Bar/Slider/keyboard) Of Phones Produced.

import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(7,7))

Touchscreens = BestSellingP.loc[BestSellingP['Form factor']=='Touchscreen'].count()[0]
Bars = BestSellingP.loc[BestSellingP['Form factor']=='Bar'].count()[0]
Flipphones = BestSellingP.loc[BestSellingP['Form factor']=='Flip phone'].count()[0]
Sliders = BestSellingP.loc[BestSellingP['Form factor']=='Slider'].count()[0]
KeyBoard_Phones = BestSellingP.loc[BestSellingP['Form factor']=='keyboard'].count()[0]

labels=["Touchscreens","Bars","Flipphones","Sliders","Keyboard Phones"]

colors=['k','m','y','g','b']
explode=[.1,.3,.2,.1,.2]
plt.title('Form Factors Of Best Selling Phones',rotation = 360)

plt.pie([Touchscreens,Bars,Flipphones,Sliders,KeyBoard_Phones],labels=labels,colors=colors,autopct='%.1f%%',explode=explode)
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```
#Importing Top Spotify Songs Of 2017 and 2018.
Spotify_songs = pd.read_csv("SpotifyTopSongs.csv")
#Since this dataset is huge, we print the top 20.

Spotify_songs.head(20)


```

```
#Finding the top 5 songs with most streams.

Songsandstreams = Spotify_songs.groupby('Track Name')['Streams'].sum() 
#Summing up songs with the most streams.abs
Songsandstreams.sort_values(ascending=False, inplace=True)
#Specifying the order of the songs by streams.

#Using IndexBasedSelection to select the top 5 most streamed songs
Songsandstreams[:5]
```

```
MostStreamed_5 = Songsandstreams[:5]
```

```
#Plotting A Bar Chart For The Most Streamed 5 Songs

plt.figure(figsize=(10,10))
colors = 'b'
width = 0.8
plt.bar(x=MostStreamed_5.index,height=MostStreamed_5,color=colors,width=width)
#x.index is used to determine the positioning of the top 5 streamed songs.
plt.title('TOP 5 STREAMED SONGS')
plt.xlabel('SONGS')
plt.ylabel('STREAMS INDEX')

plt.show()

#ED SHEERAN'S song 'Shape Of You' was the most streamed song in 2017,followed
#by the Despacito Remix by Justin Bieber,Daddy Yankee & Luis Fonsi.
```

```
#Artists With The Most Streams.
Artistsandstreams = Spotify_songs.groupby('Artist')['Streams'].sum()
Artistsandstreams.sort_values(ascending=False, inplace=True)

#10 Artists With The Most Streams.
Artistsandstreams[:10]

#The artist with the most streamed song of all was once again Ed Sheeran, followed closely by Drake.
```
