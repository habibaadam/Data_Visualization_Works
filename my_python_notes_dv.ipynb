```python
#While Loops repeats a targeted statement if the condition is met
count = 0
while (count < 10):
   print('The count is:',count)
   count = count + 1            #this means the code repeatedly runs the loop, after adding 1 to the original count,as long as it is amounts to any number less than 10
```

```python
#StackedBarCharts
#data for 2016 olympics
#plt_label is used to assign figures next to each bar charts,just so any observer would easily understand
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
x=['USA','GHana','China','Germany','Russia']
gold=[46,27,26,19,17]
bronze=[38,17,26,19,15]
silver=[37,23,18,18,10]
a=plt.bar(x,gold,label='gold',color='m')
bottom_b=[a+b for a,b in zip(gold,silver)]
b=plt.bar(x,bronze,bottom=bottom_b,label='bronze',color='g')
c=plt.bar(x,silver,bottom=gold,label='silver',color='r')
plt.bar_label(a,label_type='center')
plt.bar_label(b,label_type='center')
plt.bar_label(c,label_type='center')
plt.legend()
plt.title('Olympic Medals By Countries')
plt.ylabel('Medals')
plt.xlabel('Countries')
plt.show()
```
